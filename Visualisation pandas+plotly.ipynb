```python
#no=[1,2,1,2,4,5,8,10,11,2,3,4,5,2,110,111]             #using IQR formula to calc IQR but we can see in next box
#no=[1,2,1,2,4,5,8,10,11,2,3,4,5,2,110,111]        # that scipy stats give different value, this is coz of it is optimized 
no=[4,4,10,11,15,7,14,12,6]
a=sorted(no)
print(a)
if len(a)%2==0:
    x=(len(a)//2)
    #print(x)
    if x%2==0:
        y=x//2
        #print(a[y-1],a[y])
        #print(a[-y-1],a[-y])
        iqr=(a[-y-1]+a[-y])/2-(a[y-1]+a[y])/2
        print(iqr)
    elif x%2!=0:
        y=x//2
        #print(a[y],a[-y-1])
        iqr=a[-y-1]-a[y]
        print(iqr)
else:
    a.pop(len(a)//2)
    print(a)
    x=(len(a)//2)
    #print(x)
    if x%2==0:
        y=x//2
        #print(a[y-1],a[y])
        #print(a[-y-1],a[-y])
        iqr=(a[-y-1]+a[-y])/2-(a[y-1]+a[y])/2
        print(iqr)
        
    elif x%2!=0:
        y=x//2
        #print(a[y],a[-y-1])
        iqr=a[-y-1]-a[y]
        print(iqr)
```
