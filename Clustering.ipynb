```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("Mall_Customers.csv")
```

```python
df
```

```python
df.drop(labels=['CustomerID'],axis=1)
```

```python
x= df.iloc[:,3:]
```

```python
x
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,15):     #calc for 15 clusters
    kmean = KMeans(n_clusters=i, init='k-means++', random_state=30)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,15),wcss)            #elbow plot
```

```python
# we can see 5 clusters wll perform  well
```

```python
kmean1= KMeans(n_clusters=5, init='k-means++', random_state=30)       #telling which of out ds belongs to which cluster
kmean1.fit_predict(x)
```

```python
#now adding the cluster no to the df

x['Cluster_no']=kmean1.fit_predict(x)
```

```python
x
```

```python
x[x['Cluster_no']== 4]
```

```python
x[x['Cluster_no']== 3]
```

```python
#predicting new points cluster
kmean1.predict([[55,31]])             #shows cluster 1
```

```python
kmean1.predict([[120,70]]) 
```

```python
x=x.drop(labels=['Cluster_no'],axis=1)
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)   #eps and min are hyperparameters, so we will fig out best value ourselves
dbscan.fit(x)
```

```python
dbscan.labels_      #shows the clusters formed and the noises     #here it looks like only  1 cluster is created

#only -1 shows noises
```

```python
dbscan = DBSCAN(eps=2, min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_                     #we can see multiple clusters is created                           
```

```python
#total no of clusters 
set(dbscan.labels_)

#3 clusters- 0,1,2
```

```python
x['cluster number']= dbscan.labels_
```

```python
x
```
